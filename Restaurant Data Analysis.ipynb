Now, let's look at these three restaurants - Los Comales Taqueria, Ichiban at Austin, and Polvo's Restaurant & Bar.

```python
follow_ups = df[df['Process Description']=='1st Follow Up to Routine of 49 or less']['Restaurant Name']
```

```python
df[df['Restaurant Name'].isin(follow_ups)]
```

Here, what I have done is define the series "follow-ups" to be the set of those three restaurant names, and then pulled all associated records in the dataset. We can sort alphabetically:

```python
df[df['Restaurant Name'].isin(follow_ups)].sort_values(['Restaurant Name'])
```

Or by date:

```python
df[df['Restaurant Name'].isin(follow_ups)].sort_values(['Inspection Date'])
```

But wait! Because the data type of Inspection Date is a string, we've sorted the dates alphabetically instead of in chronological order. We need to convert the Inspection Date to dates. I'll write a function to convert the date strings to dates, and then I'll apply that function for every row in small_df.

```python
from datetime import datetime

def change_date_string_to_date(date_string: str) -> datetime:
    date = datetime.strptime(date_string, "%m/%d/%Y")
    return date
```

```python
df.sort_values(['Inspection Date'])
```

```python
df['Inspection Date'] = df['Inspection Date'].apply(lambda d: change_date_string_to_date(d))
```

```python
df.sort_values(['Inspection Date'])
```

```python
df[df['Restaurant Name'].isin(follow_ups)].sort_values(['Inspection Date'])
```

That works! Finally, I'll save this subset of our dataframe (records only associated with these three restaurants) as its own object, and create a plot using another popular plotting package in Python, called Seaborn: https://seaborn.pydata.org/index.html

```python
import seaborn as sns

three_restaurants = df[df['Restaurant Name'].isin(follow_ups)]

sns.lineplot(data=three_restaurants, x='Inspection Date', y='Score', hue='Restaurant Name')
plt.legend()
xt = plt.xticks(rotation=45)
```
